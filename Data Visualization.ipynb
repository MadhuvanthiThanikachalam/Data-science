Bar plot

```python
happy['Regional indicator'].value_counts(ascending=False).plot(kind='barh',title='Countries in each region',color='seagreen');
```

Histogram

```python
sns.histplot(happy['Logged GDP per capita'],kde=True,bins=35);
```

```python
happy["Total_score"]=(happy["Logged GDP per capita"]+happy["Social support"]+happy["Healthy life expectancy"]+happy["Freedom to make life choices"]+happy["Generosity"]-happy["Perceptions of corruption"])
```

```python
regions = happy.groupby('Regional indicator').mean().sort_values('Total_score', ascending=False).reset_index()
```

Heatmap

```python
plt.figure(figsize=(15,10))
sub = np.triu(happy.corr())
sns.heatmap(happy.corr(),linewidths=0.25,vmax=1.0, square=True, cmap="coolwarm", linecolor='yellow', annot=True,linewidth=.15, mask=sub)
```

Line Plot

```python
plt.figure(figsize=(15,5))
x= happy.groupby("Regional indicator")["Healthy life expectancy"].mean()
ax= sns.lineplot(x.index,x.values)
ax.set_ylabel("Freedom to make life choices")
plt.show()
```

Scatter plot

```python
e = happy['Social support']
f = happy['Ladder score']
plt.scatter(e, f)
plt.xlabel('Social support', weight='bold')
plt.ylabel('Ladder score', weight='bold')
plt.title('Social supports\'s correlation to hapiness score', weight='bold')
plt.show()
```

```python
from bubbly.bubbly import bubbleplot
import warnings
warnings.filterwarnings('ignore')

figure = bubbleplot(dataset = happy, x_column = 'Ladder score', y_column = 'Generosity', 
    bubble_column = 'Country name', size_column = 'Logged GDP per capita', color_column = 'Regional indicator', 
    x_title = "Ladder score", y_title = "Generosity", title = 'Happiness Score vs Generosity',
    x_logscale = False, scale_bubble = 1, height = 650)

py.iplot(figure, config={'scrollzoom': True})
```

Pie Chart

```python
d2021 = happy['Regional indicator'].value_counts()

label_d2021 = d2021.index
size_d2021 = d2021.values


colors = ['aqua', 'gold', 'yellow', 'crimson', 'magenta']

trace = go.Pie(
         labels = label_d2021, values = size_d2021, marker = dict(colors = colors), name = '2015', hole = 0.3)

data = [trace]

layout1 = go.Layout(
           title = 'Regions')

fig = go.Figure(data = data, layout = layout1)
py.iplot(fig)
```

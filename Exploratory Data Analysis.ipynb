Correlation Analysis

```python
happy.rename( columns = {
    'Overall rank' : 'rank',
    'Country or region' : 'country',
    'Score' : 'score',
    'GDP per capita' : 'gdp',
    'Social support' : 'support',
    'Healthy life expectancy' : 'expectancy',
    'Freedom to make life choices' : 'freedom',
    'Generosity' : 'generosity',
    'Perceptions of corruption' : 'corruption'
}, inplace = True)
```

```python
def plot(x, y, label1, label2):
    
    fig, ax = plt.subplots()
    ax.scatter(x, y)
    
    title = f'Correlation between {label1} and {label2}'

    ax.set(
        xlabel=label1,
        ylabel=label2,
        title=title
    )
    
    m, b = np.polyfit(x, y, 1)
    
    coeff = np.corrcoef(x,y)[0][1]
    
    print("Correlation = {0:.2f}".format(coeff))
    print()
    plt.plot(x, m*x + b)
    plt.show()
```

```python
plot(happy['gdp'],happy['score'], 'gdp', 'score')
```

```python
plot(happy['support'], happy['score'], 'support', 'score')
```

```python
plot(happy['expectancy'], happy['score'], 'expectancy', 'score')
```

Dimensionality Reduction

```python
from sklearn.decomposition import PCA 
from sklearn.cluster import KMeans 
```

```python
myset = set(happy['country'])
print(myset)
```

```python
i=0
for region in myset:
    happy['country'][happy['country'] == region] = i
    i=i+1
    
a= happy['country'] == 'India'
```

```python
X = happy.values
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

```python
mean_vec = np.mean(X_std,axis=0)
cov_mat = np.cov(X_std.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print(eig_vals)
```

```python
# Create a list of (eigenvalue, eigenvector) tuples
eig_pairs = [ (np.abs(eig_vals[i]),eig_vecs[:,i]) for i in range(len(eig_vals))]
#print(eig_pairs)
# Sort from high to low
eig_pairs.sort(key = lambda x: x[0], reverse= True)

# Calculation of Explained Variance from the eigenvalues
tot = sum(eig_vals)

var_exp = [(i/tot)*100 for i in sorted(eig_vals, reverse=True)] # Individual explained variance
cum_var_exp = np.cumsum(var_exp) # Cumulative explained variance
print(len(cum_var_exp))
```

```python
plt.step(range(9), cum_var_exp, where='mid',label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
```

```python
pca = PCA(n_components=4)
x_4d = pca.fit_transform(X_std)
print(x_4d)
```

```python
plt.figure(figsize=(9,7))
plt.scatter(x_4d[:,0],x_4d[:,1], c='goldenrod',alpha=0.5)
plt.ylim(-10,30)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3)

X_clustered = kmeans.fit_predict(x_4d)
LABEL_COLOR_MAP = {0 : 'r',1 : 'g',2 : 'b'}
label_color = [LABEL_COLOR_MAP[l] for l in X_clustered]

# Plot the scatter digram
plt.figure(figsize = (7,7))
plt.scatter(x_4d[:,0],x_4d[:,3], c= label_color, alpha=0.5) 
plt.show()
```

```python
df = pd.DataFrame(x_4d)
df['X_cluster'] = X_clustered
print(df['X_cluster'])
```

```python
sns.pairplot(df, hue='X_cluster')
```
